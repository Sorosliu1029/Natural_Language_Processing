- Elinor和Marianne出现贯穿全书，应该是主角的身份
- Edward和Willoughby交替出现，时间轴上基本没有交集，但是时间轴的并集却是贯穿全书的，可见两人是很重要的角色
- 在不了解内容的情况下，猜测Edward和Willoughby在地理上是分开的角色，（或是同一人用不同的的名字）
